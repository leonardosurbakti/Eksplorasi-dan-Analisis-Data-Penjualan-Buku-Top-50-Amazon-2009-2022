## Mengimport library dan Membaca File .csv

```python
# Import library yang dibutuhkan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('bestsellers_with_categories_2022_03_27.csv')
data.head()
```

## Exploratory Data Analysis

```python
data.info()
```

```python
data.shape
```

```python
data['Genre'].unique()
```

```python
plt.figure(figsize = (16,7))
sns.lineplot(x="Year", y="User Rating", hue="Genre", data=data).set(title='User Rating Berdasarkan Genre dari Tahun 2009-2022')
```

Terlihat bahwa ada fluktuasi pada jumlah user rating untuk buku best seller di genre fiksi dan nonfiksi. Namun pada tahun 2019 sampai tahun 2021 terjadi kenaikan jumlah user rating. Pada 2019, khususnya, terjadi kenaikan jumlah user rating pada genre fiksi. Kondisi ini dapat diakibatkan karena adanya pandemi COVID-19 yang mengharuskan setiap orang untuk tetap di rumah yang dapat mendorong orang lebih banyak membaca buku sehingga jumlah user rating meningkat jumlahnya.

```python
plt.figure(figsize=(16,7))
sns.boxplot(x="Year", y="Reviews", hue="Genre", palette=["m", "g"], data=data)
```

Jumlah review meningkat secara drastis dari tahun 2020 untuk buku fiksi dan nonfiksi. Grafik ini dapat menunjukkan bahwa banyak orang kembali membaca buku selama pandemi covid-19. Aturan pemerintah yang melarang orang untuk berkumpul dan budaya bekerja dari rumah (WFH) mendorong orang untuk lebih menghabiskan waktunya untuk membaca buku. Buku fiksi terlihat lebih populer dibanding buku nonfiksi, namun buku nonfiksi juga tetap diminati pembaca, terlihat dari jumlah review yang hampir menyamai buku fiksi.

```python
plt.figure(figsize = (16,7))
sns.lineplot(x="Year", y="Price", hue="Genre", data=data).set(title='Harga Buku Berdasarkan Genre dari Tahun 2009-2022')
```

Faktor lain yang dapat menjelaskan kenapa terjadi peningkatan jumlah user rating dan reviews di Amazon adalah karena harga buku fiksi pada tahun 2019 - 2021 tercatat mengalami penurunan, sedangkan untuk buku nonfiksi mengalami kenaikan harga dari mulai dari tahun 2019, namun terlihat bahwa buku (top 50) bergenre fiksi dan nonfiksi mengalami penurunan harga yang signifikan di website Amazon. Penurunan harga buku best seller ini juga dapat terjadi akibat efek pandemi Covid-19 yang mendorong orang untuk lebih banyak membaca buku karena harus diam di rumah sehingga terjadi lonjakan review dan rating dari pembaca.

```python
plt.figure(figsize=(16,7))
sns.scatterplot(x='User Rating', y='Reviews', data=data).set(title="Sebaran Jumlah Review berdasarkan User Rating")
```

```python
df=data.groupby(['Year','Genre']).size().reset_index(name='Jumlah buku')
plt.figure(figsize=(16,7))
sns.barplot(x='Year', y='Jumlah buku', hue='Genre', data=df).set(title='Jumlah buku fiksi vs nonfiksi dalam Amazon top-50 list tahun 2009-2022')
```
